```python
# removing the outlayers from the data set:

q1 = df.drop(columns = "Fuel_Type").quantile(0.25)
q3 = df.drop(columns = "Fuel_Type").quantile(0.75)
iqr = q3-q1
df = df[
        ~((df.drop(columns = "Fuel_Type").lt(q1 - 1.5 * iqr)) | (df.drop(columns = "Fuel_Type").gt(q3 + 1.5 * iqr))).any(axis=1)
]
```

```python
plt.boxplot(df.drop(columns = "Fuel_Type"))
```

```python
df.info()
```

```python
# data transformation :

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

ss = StandardScaler()
le = LabelEncoder()
df["Fuel_Type"] = le.fit_transform(df["Fuel_Type"])
df.info()
```

```python
df.head()
```

```python
# deviding the dataset into the x and y variables:

X = df.iloc[:,1:]
Y = df["Price"]

# standardization of the x variables:

X = ss.fit_transform(X.iloc[:,:])
X
```

```python
# splitting the data into train and test dataset:

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
# model fitting with the linear regression:

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
model = LinearRegression()
model.fit(x_train,y_train)
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

train_acc = r2_score(y_train,y_train_pred)
test_acc = r2_score(y_test,y_test_pred)
print("train_accuracy :",np.round(train_acc,2))
print("test_accuracy :",np.round(test_acc,2))


```

yaa!..
model fitted with the linear regression which i have got the results like this
train_accuracy : 0.81
test_accuracy : 0.81

i have got both the testing and training as same results that is 81%.
and this is tried without the cross validation.

```python
# calculating the vif csore for the variables:

from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

p = df.iloc[:, 1:].copy()
p = add_constant(p)

vif_df = pd.DataFrame()
vif_df["Feature"] = p.columns
vif_df["VIF"] = [variance_inflation_factor(p.values, i) for i in range(p.shape[1])]

print(vif_df)
```

```python
# cross validation:

training_accuracy = []
test_accuracy = []
for i in range(1,201):
    x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size= 0.3, random_state= i)
    model.fit(x_train,y_train)
    y_train_predict = model.predict(x_train)
    y_test_predict = model.predict(x_test)
    training_accuracy . append(r2_score(y_train,y_train_predict))
    test_accuracy .append(r2_score(y_test,y_test_predict))
print("Cross validaton training accuracy:",np.round(np.mean(training_accuracy)*100,2))
print("Cross validaton test accuracy:",np.round(np.mean(test_accuracy)*100,2)) 
```

here are the results with the cross validation and the model is linear regression
and the results are like this
    Cross validaton training accuracy: 81.44
    Cross validaton test accuracy: 80.87


not much difference  in the results with the cross validation with the same model that is linear regression

2. Multicolinearity issues are nothing but both the independent x variables are having the high corelation between them rather than having the high        corelation with the target variable is known as Multicolinearity issues .
   Techniques are:

   1.calculating the VIF(variance inflation factor):
     -> if the features which are having the high vif_score,then we can remove that features for the better performance.
   2.Regularization( ridge and lasso):
      with this regularization techniques also we can reduce the Multicolinearity issues by removing the unwanted featured from the model fitting
   3.PCA(principle conponent analysis):
     this technique used to deduce the dimentionality of the data set.
     which will removes the unwanted features which are having no relation with the taget variable.
