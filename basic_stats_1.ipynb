```python
# Bar plots for categorical columns:
# for bar plot we have to import the following:
import matplotlib.pyplot as plt

#collecting the all catogorical columns into list:
catogorical_col = df.select_dtypes(include = "object").columns.tolist()

# for calling the single column at a time from the above list:
for col in catogorical_col:
    df[col].value_counts().plot(kind="bar")
    plt.ylabel("count")
    plt.show()
```

```python
#collecting the all numerical columns into a single list.

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# for standardization we have to import the following method
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
standardized_data = ss.fit_transform(df[numerical_cols])# here we are standardizing all the numerical columns at a time since all are in the same list:

#converting the standardized_data into the dataframe:
standardized_df = pd.DataFrame(standardized_data, columns=[col for col in numerical_cols])
print("before standardization:",df[numerical_cols].head())
print("after standardization: ",standardized_df.head())
```

```python
#one-hot encoding in data analysis and machine learning
for col in catogorical_col:
    dummy = df[col] == False
    print(dummy)
```

```python
df_encoded = pd.get_dummies(df[catogorical_col], drop_first=True)
df_encoded
```

```python
final_df = pd.concat([standardized_df, df_encoded], axis=1)
final_df
```
