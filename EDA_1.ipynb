```python
# filling the null values with the mean and medians :

for col in ['AC','DS','DP','Tendency','NSP']:
    df[col].fillna(df[col].mean(),inplace=True)
for col in ['LB','MLTV','Width']:
    df[col].fillna(df[col].median(),inplace=True)
df.isnull().sum()  
```

```python
# removeing the outliers in the dataset:

q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3-q1
df = df[
    ~((df.lt(q1 - 1.5 * iqr)) | (df.gt(q3 + 1.5 * iqr))).any(axis=1)
]
```
